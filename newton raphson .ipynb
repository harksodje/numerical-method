#Newton Raphson Numerical Method

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# initialization
# function of x 
def fn(x):
    return np.cos(x)
#derivative of function
def df(x):
    return -np.sin(x)
#initial guess of root
x0 = 2
# actual root
actual = np.pi/2
# max number of iteration
maxit = 20
# min divisor allow to not divide by too small if so we need to break
minidivisor = 1E-20
#tolerance for error
tolerance = 5E-8
#array to store data for each iteration
approxerror =np.zeros(maxit)
actualerror = np.zeros(maxit)
approxrelerror = np.zeros(maxit)
actualrelerror = np.zeros(maxit)

## calculations
xi = x0
# loop
for i in range (maxit):
    #calculate actual and relative errors
    actualerror[i] = abs(actual - xi)
    #print (actual)
    actualrelerror[i] = abs(actualerror[i]/actual)*100
    fnprime = df(xi)
    if abs(fnprime) < minidivisor:
        print ("derivative is too small")
    # update root estimate (xi plus 1)
    xii = xi - fn(xi)/fnprime
    
    approxerror[i+1] = abs (xii-xi)
    approxrelerror[i+1] = abs (approxerror[i+1]/xii)
    if approxrelerror[i+1 ] <= tolerance:
        break
    #update
    xi = xii
# presentation
# print root
print ("root at: ", xi, "with tolerance ", tolerance)
plt.figure(1)
plt.plot (np.arange(1,i+1),approxrelerror[1:i+1], label = "approx. rel. error")
plt.plot (np.arange(i+1),actualrelerror[:i+1], label = "actual. rel. error")
plt.ylabel('relative errors ')
plt.xlabel('iteration number')
plt.yscale('log')
plt.legend()
plt.show()
print (approxrelerror)
print (actualrelerror)
```
